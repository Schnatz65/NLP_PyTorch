```python
import requests
from zipfile import ZipFile

# Der Dowmload dauert u.U. mehrere Minuten, da Ã¼ber 800 MB "gezogen" werden.
url = 'http://nlp.stanford.edu/data/glove.6B.zip'
print("Datei holen ...")
r = requests.get(url, allow_redirects=True)
print("Datei schreiben ...")
open('data/glove.6B.zip', 'wb').write(r.content)
print("Dateien unzippen...")
with ZipFile('data/glove.6B.zip', 'r') as zipObj:
   zipObj.extractall(path="data/glove.6B")
print("Fertig!")
```
